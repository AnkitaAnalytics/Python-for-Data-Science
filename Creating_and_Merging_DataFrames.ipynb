```python
import numpy as np
import pandas as pd
```

### Creating a Dataframe from list

```python
cust_data = [['001', 'Alice', 'alice@example.com', '123-456-7890', 20],
             ['002', 'Bhooshan', 'bob@example.com', '987-654-3210', 15],
             ['003', 'Carl', 'charlie@example.com', '555-123-4567', 25]]

cust_cols = ['CustomerID', 'Name', 'Email', 'Phone', 'TotalPurchases']
# Creating a dataframe
df = pd.DataFrame(data = cust_data, columns = cust_cols)

print(df)
```

```python
# Creating Index

cust_index = ["A","B","C"]
df=pd.DataFrame(data=cust_data,columns = cust_cols,index=cust_index)

print(df)
```

### Creating DataFrames from Dictionaries

```python
cust_data= {"CustomerID":["001","002","003"],"Name":["Alice","Bhooshan","Carl"],
            "Email":["alice@example.com","bob@example.com","carl@example.com"],
            "Phone":["123-456-7890","987-654-3210","555-123-4567 "],
            "TotalPurchases":[20,15,25]}

df=pd.DataFrame(data=cust_data) #the column names are inferred from the keys of the dictionary
print(df)
```
