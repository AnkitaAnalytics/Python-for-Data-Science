### Create a panda series that contains the first ‘n’ natural numbers and their respective squares. The first ‘n’ numbers should appear in the index position.

```python
n= int(input("Enter n "))
```

```python
#creating a dictionary with numbers as keys (manual indexing) and their squares as values
data= {i:i**2 for i in range (1,n+1)}
squares = pd.Series(data)

print(squares) #printing the seires
```

### Indexing Dataframes
Print only the even numbers of rows of the dataframe 'df'.
Note: Don't include the row indexed zero.

```python
# changing the sqaures series into a dataframe
col= ["Squares"]
df= pd.DataFrame(squares,columns=col)
print(df)
```

```python
# printing the even rows
df2= df.iloc[1::2]
print(df2)
```

### Group the dataframe 'forestfires' by 'month' and 'day' and find the mean value for columns 'rain' and 'wind'
df link= https://query.data.world/s/vBDCsoHCytUSLKkLvq851k2b8JOCkF


```python
# reading the dataset and understanding it
ff= pd.read_csv("https://query.data.world/s/vBDCsoHCytUSLKkLvq851k2b8JOCkF")

print(f"Shape of dataset {ff.shape}")
ff.head(3)
```

```python
ff1= ff.groupby(["month","day"])[["rain","wind"]].mean()
print(ff1.head(10))
```

```python
# using agg()- it allows to specify operations for each column

ff2 = ff.groupby(["month","day"]).agg({"rain":"mean","wind":"mean"})
print(ff2.head(10))
```

```python
# Without Selecting Columns Before Grouping

ff3= ff.groupby(["month","day"]).mean()[["rain","wind"]]
ff3.head(10)
```

```python
# finding mean of rain and wind using pivot table command
df_1= ff.pivot_table(
    values=["rain","wind"], # Columns to aggregate
    index=["month","day"],  # Columns to group
    aggfunc="mean")         #function to perform

df_1.head(10)
```
