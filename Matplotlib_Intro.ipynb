```python
# importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data= {"C":250,"C++":430,"Python":440,"Java":180}
courses = list(data.keys()) #storing the course name in a list
students= list(data.values())

#Ploting this on bar graph
plt.bar(courses,students)
```

```python
# changing the size of the figure

fig= plt.figure(figsize= (5,5))

# x - label name
plt.xlabel("Languages")
plt.ylabel("Students")
plt.title("Students vs Courses")
plt.xticks(rotation = 45)
plt.yticks(rotation = 45)

plt.bar(courses,students,color ="green", align="center",edgecolor="black")


```

```python
# Plotting
fig = plt.figure(figsize=(5, 5))

plt.xlabel("Languages")
plt.ylabel("Students")
plt.title("Students vs Courses")
plt.xticks(rotation=45)
plt.yticks(rotation=0)  # y-axis labels don't need rotation

# Plot bars
bars = plt.bar(courses, students)

# Add annotations
for bar in bars:
    yval = bar.get_height()  # Get the height of the bar
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # x-coordinate
        yval + 1,  # y-coordinate (slightly above the bar)
        f"{yval}",  # Text to display (the height value)
        ha="center",  # Center align
        va="bottom"   # Bottom align
    )

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example data
data = {
    "Courses": ["Python", "Java", "C++", "JavaScript", "Ruby"],
    "Students": [45, 30, 50, 20, 35]
}
df = pd.DataFrame(data)

# Sorting directly in Seaborn
plt.figure(figsize=(6, 5))
sns.barplot(
    data=df,
    x="Courses",
    y="Students",
    order=df.sort_values("Students")["Courses"],  # Sort courses by student count
)

# Adding labels and title
plt.xlabel("Languages")
plt.ylabel("Students")
plt.title("Students vs Courses (Sorted)")

# Adding annotations
for index, row in df.sort_values("Students").iterrows():
    plt.text(
        x=index,
        y=row["Students"] + 1,
        s=row["Students"],
        ha="center",
        va="bottom"
    )

plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example data
data = {
    "Courses": ["Python", "Java", "C++", "JavaScript", "Ruby"],
    "Students": [45, 30, 50, 20, 35]
}
df = pd.DataFrame(data)

# Sorting the data for the plot
sorted_df = df.sort_values("Students")

# Create the bar plot
plt.figure(figsize=(6, 5))
barplot = sns.barplot(
    data=sorted_df,
    x="Courses",
    y="Students",
    palette="viridis"
)

# Add annotations directly
for bar, student_count in zip(barplot.patches, sorted_df["Students"]):
    bar_height = bar.get_height()  # Get the height of each bar
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # Center text horizontally on the bar
        bar_height + 1,  # Position text slightly above the bar
        f"{student_count}",  # Display the student count
        ha="center",  # Horizontal alignment
        va="bottom"   # Vertical alignment
    )

# Add labels and title
plt.xlabel("Languages")
plt
```
