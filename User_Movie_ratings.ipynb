```python
import pandas as pd
```

```python
path ="/content/drive/MyDrive/Datasets/User_Movie_Ratings (1).csv"
df_movies=pd.read_csv(path)
```

### Retrieve the rating ‘Indira’ gave to the movie ‘Annabelle’ in the df_movies dataframe

```python
df_movies = df_movies.set_index('User') #changing the index as user
ratings= df_movies.loc['Indira', 'Annabelle'] #getting the ratings at index Indira and column Annabelle
print(ratings)
```

### Find the ratings given by ‘Manish’ and ‘Ravi’ for the movies ‘Thor’, ‘Justice League’, and ‘Spiderman’ in the df_movies dataframe.

```python
df_movies.loc[['Manish', 'Ravi'], 'Thor':'Spiderman']
```

### How many users in the df_movies dataframe rated more than 7 for Thor and Spiderman but rated 7 or less for Dolittle. How will you extract this information from the dataframe?

```python
df_movies[(df_movies['Thor'] > 7) & (df_movies['Spiderman'] > 7) & (df_movies['Dolittle'] <= 7)].shape[0]
```

### Find the mean action movie rating for each user in the dataframe df_movies without modifying the original dataframe and identify users whose mean action movie ratings are more than 8.


```python
#creating a pandas series named action_movies
action_movies = (df_movies['Thor'] + df_movies['Justice League'] + df_movies['Spiderman']) / 3
action_movies[action_movies > 8].index
```

```python
df_movies.describe().transpose()['mean'].idxmax()
```

### Calculate the average rating given to each movie

```python
df_movies.mean(axis=0)
```

### Replace movie ratings above 5 with ‘good’ and everything else with ‘bad’ without modifying the original dataframe in place.

```python
def corrections(row):
    return 'good' if row > 5 else 'bad'
new_df = df_movies.applymap(corrections)
```
