```python
import numpy as np
import pandas as pd
```

### Creating a Dataframe from list

```python
cust_data = [['001', 'Alice', 'alice@example.com', '123-456-7890', 20],
             ['002', 'Bhooshan', 'bob@example.com', '987-654-3210', 15],
             ['003', 'Carl', 'charlie@example.com', '555-123-4567', 25]]

cust_cols = ['CustomerID', 'Name', 'Email', 'Phone', 'TotalPurchases']
# Creating a dataframe
df = pd.DataFrame(data = cust_data, columns = cust_cols)

print(df)
```

```python
# Creating Index

cust_index = ["A","B","C"]
df=pd.DataFrame(data=cust_data,columns = cust_cols,index=cust_index)

print(df)
```

### Creating DataFrames from Dictionaries

```python
cust_data= {"CustomerID":["001","002","003"],"Name":["Alice","Bhooshan","Carl"],
            "Email":["alice@example.com","bob@example.com","carl@example.com"],
            "Phone":["123-456-7890","987-654-3210","555-123-4567 "],
            "TotalPurchases":[20,15,25]}

df=pd.DataFrame(data=cust_data) #the column names are inferred from the keys of the dictionary
print(df)
```

### Changing index using set_index

```python
df.set_index("CustomerID", inplace= True ) # using the values of an existing column as the index
```

```python
print(df)
```

###  Conditional (boolean) indexing

```python
#printing first 2 rows to know about the data
df.head(2)
```

```python
# lets see Order where profit is greater than 200

df["Profit"]>200 # this prints False where profit <200 and True for all rows where profit>200
```

```python
#accessing rows where profit >200

df[df["Profit"]>200]
```

```python
# combining 2 or more than 2 conditions

df[(df["Profit"]>200) & (df["Category"]=="Office Supplies")] #gives the data where profit is greater than 200 for category "Office Supplies"
```
