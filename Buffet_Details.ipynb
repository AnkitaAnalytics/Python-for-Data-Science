```python
import pandas as pd
```

```python
buffet_data = [['A1', 'Shilpa',18,'Indian',24.65],
               ['A2','Jaspreet',32,'American',18.54],
               ['A3','Dominic',19,'Indian',	27.66],
               ['A4','Ahmad',22,'American',19.54],
               ['A5','Joseph',28,'Indian',17.32],]
buffet_cols=['Room',	'Name',	'Age',	'Cuisine',	'Expenditure']
df_buffet = pd.DataFrame(data=buffet_data,columns=buffet_cols)
```

```python
df_buffet.head()
```

### Access the 'Cuisine' column in the dataframe

```python
#method 1 - square bracket notation is used to access a specific column in Pandas
df_buffet['Cuisine']
```

```python
# method 2- get() method is used to access a single column in Pandas
df_buffet.get('Cuisine')
```

### Find the average age of customers

```python
df_buffet['Age'].mean()
```

### filter the dataframe to include only rows where the 'Expenditure' is less than 20

```python

df_buffet[df_buffet['Expenditure'] < 20]
```
